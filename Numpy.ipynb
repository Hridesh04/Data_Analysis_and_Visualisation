Plotting Trigometric Curves

```python
import matplotlib.pyplot as plt
x_sin = np.arange(0,2*np.pi,0.1)
y_sin = np.sin(x_sin)

plt.figure(figsize=(6, 6))
plt.plot(x_sin, y_sin)
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x_cos = np.arange(0,2*np.pi,0.1)
y_cos = np.cos(x_cos)

plt.figure(figsize=(6, 6))
plt.plot(x_cos, y_cos)
plt.title('Cosine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x_tan = np.arange(0,2*np.pi,0.1)
y_tan = np.tan(x_tan)

plt.figure(figsize=(6, 6))
plt.plot(x_tan, y_tan)
plt.title('Tangent Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x_cot = np.arange(0,2*np.pi,0.1)
y_cot = 1/np.tan(x_cot)

plt.figure(figsize=(6, 6))
plt.plot(x_cot, y_cot)
plt.title('Cotangent Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# print all in a single plot
plt.figure(figsize=(6, 6))
plt.plot(x_sin, y_sin, 'g--', label='Sine Wave')
plt.plot(x_cos, y_cos, 'b--', label='Cosine Wave')
plt.plot(x_tan, y_tan, 'r--', label='Tangent Wave')
plt.plot(x_cot, y_cot, 'm--', label='Cotangent Wave')
plt.title('Trigonometric Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
# print all in a different plot
plt.figure(figsize=(6, 6))
plt.subplot(2, 2, 1)
plt.plot(x_sin, y_sin, 'g-')
plt.title('Sine Wave')

plt.subplot(2, 2, 2)
plt.plot(x_cos, y_cos, 'b-')
plt.title('Cosine Wave')

plt.subplot(2, 2, 3)
plt.plot(x_tan, y_tan, 'r-')
plt.title('Tangent Wave')

plt.subplot(2, 2, 4)
plt.plot(x_cot, y_cot, 'm-')
plt.title('Cotangent Wave')
```
