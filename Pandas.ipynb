### Handling the Null Values

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\HRSHUKL\Downloads\sample.csv')
df.head()
```

```python
df
```

```python
df.isnull() # to check whether there is a null values available or not
```

```python
df.isnull().sum() # by column it will tell how many null values are there
```

```python
df.isnull().sum().sum()
```

# Dropping Rows with NaN values

```python
df.shape
```

```python
df2 = df.dropna() # useing this function we cna drop all the rows which are having any NULL values
```

```python
df2.shape
```

##  Dropping Columns with NaN values

```python
df3 = df.dropna(axis = 1) # axis = 1 for columns
df3.shape

```

```python
df.dropna(how='any') #if any row calue is null then remove that row
```

```python
df.dropna(how='all') #if all row value is null then remove that row
```

```python
df.dropna(inplace=True) # it will replace the values in the original data frame and return it back to us
```

```python
df.shape
```

## Filling the NULL Values

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\HRSHUKL\Downloads\sample.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(0) # all the null values are filled with zeros or we can fill with any number
```

```python
df.fillna(method='ffill')
```
